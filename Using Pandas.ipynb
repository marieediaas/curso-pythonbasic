```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_data = {"names": ["Marie", "Nick", "Neto", "Mimi", "Meire"],
           "jan_ir": [143, 122, 101, 106, 365],
           "feb_ir": [121, 222, 148, 152, 256],
           "mar_ir": [191, 120, 234, 161, 152]}

df = pd.DataFrame(raw_data, columns=["names", "jan_ir", "feb_ir", "mar_ir"])
df["total_ir"] = df["jan_ir"] + df["feb_ir"] + df["mar_ir"]

color = [(1, .4, .4), (1,.6, 1), (.5, .3, 1), (.3, 1, .5), (.7, .7, .2)]

plt.pie(df["total_ir"],
       labels = df["names"],
       colors = color,
       autopct = "%1.1f%%")
plt.show()
```
