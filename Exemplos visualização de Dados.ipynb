```python
import matplotlib.pyplot as plt

years = [1950, 1955, 1960, 1965, 1970,1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]

pop = [2.5, 2.7, 3.0, 3.3, 3.6, 4.0, 4.4, 4.8, 5.3, 5.7, 6.1, 6.5, 6.9, 7.3]

deaths = [1.2,1.7,1.8,2.2,2.5,
         2.7, 2.9, 3, 3.1, 3.3, 3.5, 3.8, 4.0, 4.3]

#utilizando dashed line
plt.plot(years,pop,'--', color = (255/255,100/255,100/255))
plt.plot(years, deaths, color = (.6,.6,1))

plt.ylabel("Population in Billions")
plt.xlabel("Years")
plt.title("Population Growth")
plt.show()
```

```python
fig = plt.figure()

# axes: esquerda, abaixo, largura, altura (valores de 0 a 1)
axes = fig.add_axes([0.,0.8,0.8,0.8])

axes.plot(years,pop)

axes.set_xlabel("Years")
axes.set_ylabel("Population in Billions")
axes.set_title("Population Growth")
```

```python
lines = plt.plot(years, pop, years,deaths)
plt.grid(True)

plt.setp(lines, color =(1,.4,.4), marker= 'o')
plt.show()
```
